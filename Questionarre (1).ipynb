```python
import ipywidgets as widgets
import pandas as pd
from IPython.display import display


# List of options for the dropdown
options = ['Not at all', 'Several days', 'More than half the days','Nearly every day']

# List to store answers corresponding to each question
answers = []

# Initialize question index to show one question at a time
question_index = 0

# Create a dropdown widget
dropdown = widgets.Dropdown(
    options=options,
    value=None,  # Set default to the first option
    description='Select:',
    disabled=False
)

# Function to handle selection and append to the list
def on_change(change):
    global question_index
    
    # Save the answer to the list
    answers.append(change.new)
    
    # Print the current question and selected answer
    print(f'Question{question_index + 1}: {df["Questions"][question_index]}')
    print(f'Selected Answer: {change.new}')
    
    question_index += 1
    
    if question_index < len(df):
        # Update the question text and options
        question_label.value = df["Questions"][question_index]
        dropdown.value = options[0]  # Reset to default
    else:
        # No more questions, save the answers to the DataFrame
        df['answers'] = answers
        print("Survey complete! Here are your answers:")
        print(df)
    # Move to the next question or finish if there are no more questions
    
# Label to show the current question
print(question_index)
question_label = widgets.Label(value=df["Questions"][question_index])

# Display the question label and dropdown
display(question_label)
display(dropdown)

# Link the dropdown to the event listener
dropdown.observe(on_change, names='value')
```
